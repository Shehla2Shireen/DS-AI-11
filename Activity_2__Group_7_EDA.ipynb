```python
# Calculate IQR
Q1 = df['Total Spent'].quantile(0.25)
Q3 = df['Total Spent'].quantile(0.75)
IQR = Q3 - Q1
# Define outlier condition
outliers = df[(df['Total Spent'] < (Q1 - 1.5 * IQR)) | (df['Total Spent'] > (Q3 + 1.5 * IQR))]
print("Number of outliers:", outliers.shape[0])
print(outliers[['Total Spent']].head())
import seaborn as sns
import matplotlib.pyplot as plt

sns.boxplot(x=df['Total Spent'])
plt.title("Boxplot of Total Spent")
plt.show()
#since we have only 1 outlier, do cappng
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
df['Total Spent'] = df['Total Spent'].clip(lower=lower_bound, upper=upper_bound)
sns.boxplot(x=df['Total Spent'])
plt.title("Boxplot of Total Spent")
plt.show()
```

**Sales Trends:**

```python
#Line plots showing sales over time.
# Step 1: Group by 'Transaction Date' and sum the 'Total Spent'
df_grouped = df.groupby('Transaction Date')['Total Spent'].sum().reset_index()

# Step 2: Plot Line Plot for Sales Over Time
plt.figure(figsize=(10, 6))
plt.plot(df_grouped['Transaction Date'], df_grouped['Total Spent'], marker='o')
plt.title('Sales Over Time')
plt.xlabel('Transaction Date')
plt.ylabel('Total Spent')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#Bar charts for sales by product category.

# Step 1: Group by 'Item' and sum 'Total Spent'
df_item_sales = df.groupby('Item')['Total Spent'].sum().reset_index()

# Step 2: Plot Bar Chart for Sales by Product Category (Item)
plt.figure(figsize=(10, 6))
plt.bar(df_item_sales['Item'], df_item_sales['Total Spent'], color='skyblue')
plt.title('Sales by Product Category')
plt.xlabel('Product Category (Item)')
plt.ylabel('Total Spent')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

**Customer Behavior:**

```python
#Pie charts depicting purchase distribution among customer segments
# Step 1: Group by 'Item' and sum 'Total Spent'
df_item_sales = df.groupby('Item')['Total Spent'].sum()

# Step 2: Plot Pie Chart
plt.figure(figsize=(8, 8))
plt.pie(df_item_sales, labels=df_item_sales.index, autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired.colors)
plt.title('Purchase Distribution Among Customer Segments (Items)')
plt.show()
```

```python
#Heatmaps to identify peak sales hours or days.


import seaborn as sns
import numpy as np
# Step 1: Extract day and hour from 'Transaction Date'
df['Day'] = df['Transaction Date'].dt.day_name()
df['Hour'] = df['Transaction Date'].dt.hour

# Step 2: Pivot the data to get sales by day and hour
heatmap_data = df.pivot_table(index='Day', columns='Hour', values='Total Spent', aggfunc=np.sum)

# Step 3: Plot Heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(heatmap_data, annot=True, fmt='.1f', cmap='YlGnBu', cbar_kws={'label': 'Total Spent'})
plt.title('Peak Sales Hours or Days')
plt.xlabel('Hour of the Day')
plt.ylabel('Day of the Week')
plt.tight_layout()
plt.show()
```
